```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame({
    "t": ["t1", "t1", "t1", "t2", "t2", "t2"],
    "m": ["m1", "m2", "m3", "m1", "m2", "m3"],
    "v": [10, 12 ,23, 4, 18, 27]
})
fig, ax = plt.subplots(1, 1, figsize=(20, 5))
sns.barplot(x="t", y="v", hue="m", data=data, orient="v", ax=ax)
ax.set_xlabel('Texts')
ax.set_ylabel('Value')
ax.set_xticklabels(["sdgsd" + "\n" + "hdfgdf", "sgdfhd"], rotation='horizontal', fontsize=15)

# alternative way to control ticks:
#     plt.xticks(rotation=90)
#     for tick in ax.get_xticklabels():
#         tick.set_rotation(45)
#         tick.set_fontsize(11)
```
